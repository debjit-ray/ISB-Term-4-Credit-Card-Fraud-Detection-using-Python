# 5. Build Different Models

### _Approach - 1_ : 

#### Considering we have a highly imbalanced dataset with very few Fraud transactions, we are going to use K-fold cross validation to train our models. Within each fold, we will use SMOTE to oversample the Fraud observations. This will ensure that our models do not get heavily biased against the Fraud transactions.

```python
from sklearn.model_selection import KFold
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

def fnc_KFold(model, mdl_name, predictors_train, actual_train):
    n_splits = 7
    kf = KFold(n_splits=n_splits)
    accuracy = []
    precision = []
    recall = []
    i = 1
    print("---------------------------------------------------------------")
    print("*                     {}                  *".format(mdl_name))
    print("---------------------------------------------------------------")
    for train_index, test_index in kf.split(predictors_train):
        X_train = predictors_train.iloc[train_index,]
        X_test = predictors_train.iloc[test_index,]
        Y_train = actual_train.iloc[train_index,]
        Y_test = actual_train.iloc[test_index,]

        # Ensure oversampling to balance the dataset before training
        smote_sampler = SMOTE(random_state=24)
        X_train_smote, Y_train_smote = smote_sampler.fit_resample(X_train, Y_train)
        # Train the model
        model.fit(X_train_smote, Y_train_smote) #Training the model

        model_prediction = model.predict(X_test)
        print(f"Accuracy for the fold no. {i} on the test set: {accuracy_score(np.array(Y_test).reshape(-1, 1), model_prediction)}")
        accuracy.append(accuracy_score(np.array(Y_test).reshape(-1, 1), model_prediction))
        precision.append(precision_score(Y_test, model_prediction))
        recall.append(recall_score(Y_test, model_prediction)  )
        i += 1   
    print ("Average Accuracy of {} is : {}".format(mdl_name, np.round(np.average(accuracy),5)))
    print ("Average Precision of {} is : {}".format(mdl_name, np.round(np.average(precision),5)))
    print ("Average Recall of {} is : {}".format(mdl_name, np.round(np.average(recall),5)))
    print("\n")
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

def fnc_TrainModel(model, mdl_name, predictors_train, actual_train):
    print("---------------------------------------------------------------")
    print("*                     {}                  *".format(mdl_name))
    print("---------------------------------------------------------------")

    # Train the model
    model.fit(predictors_train, actual_train) #Training the model

    model_prediction = model.predict(predictors_train)
    accuracy = accuracy_score(np.array(actual_train).reshape(-1, 1), model_prediction)
    precision = precision_score(actual_train, model_prediction)
    recall = recall_score(actual_train, model_prediction) 

    print ("Accuracy of {} is : {}".format(mdl_name, np.round(accuracy,5)))
    print ("Precision of {} is : {}".format(mdl_name, np.round(precision,5)))
    print ("Recall of {} is : {}".format(mdl_name, np.round(recall,5)))
    print("\n")
```

# 6. Test Different Models using Validation dataset

```python
from sklearn import metrics
def fncValidate(model, mdl_name, X, Y):
    model_prediction = model.predict(X)    
    accuracy = np.round(accuracy_score(Y, model_prediction),5)
    # Calculate precision, recall, true positive rate, false positive rate and AUC for different models        
    precision = np.round(precision_score(Y, model_prediction),5)
    recall = np.round(recall_score(Y, model_prediction),5)        
    model_prediction_prob = model.predict_proba(X)[::,1]
    fpr, tpr, _ = roc_curve(Y, model_prediction_prob)
    auc = roc_auc_score(Y, model_prediction_prob)
    cm = metrics.confusion_matrix(Y, model_prediction)
    return(accuracy, precision, recall, fpr, tpr, auc, cm )
```

#### Test all the models where we have used only selective features and store the performance metrics into a dataframe

```python
score_list = []
model_scores = pd.DataFrame(columns = ['setName','ModelName','Features', 'Approach', 'Accuracy','Precision','Recall','confmatrix_TN','confmatrix_FP','confmatrix_FN','confmatrix_TP','fpr','tpr','auc'])

for (model, mdl_name, features, approach) in (model_list_selFtrs_kFold + model_list_selFtrs_direct):
    print(mdl_name)
    dict1 = {}
    accuracy, precision, recall, fpr, tpr, auc, cm  = fncValidate(model, mdl_name, predictors_selFtrs_val, actual_val)
    dict1.update({'setName': 'Validation',
                  'ModelName': mdl_name,
                  'Features': features,
                  'Approach': approach,
                  'Accuracy': accuracy,
                  'Precision': precision,
                  'Recall': recall,
                  'confmatrix_TN': cm[0,0],
                  'confmatrix_FP': cm[0,1],
                  'confmatrix_FN': cm[1,0],
                  'confmatrix_TP': cm[1,1],
                  'fpr': [fpr],
                  'tpr': [tpr],
                  'auc': auc})

    score_list.append(dict1)
model_scores = model_scores.append(score_list, ignore_index= True)
```

#### Test all the models where we have used all the features and store the performance metrics into the same dataframe

```python
for (model, mdl_name, features, approach) in (model_list_allFtrs_kFold + model_list_allFtrs_direct):
    print(mdl_name)
    dict1 = {}
    accuracy, precision, recall, fpr, tpr, auc, cm  = fncValidate(model, mdl_name, predictors_allFtrs_val, actual_val)
    dict1.update({'setName': 'Validation',
                  'ModelName': mdl_name,
                  'Features': features,
                  'Approach': approach,
                  'Accuracy': accuracy,
                  'Precision': precision,
                  'Recall': recall,
                  'confmatrix_TN': cm[0,0],
                  'confmatrix_FP': cm[0,1],
                  'confmatrix_FN': cm[1,0],
                  'confmatrix_TP': cm[1,1],
                  'fpr': [fpr],
                  'tpr': [tpr],
                  'auc': auc})

    score_list.append(dict1)
model_scores = model_scores.append(score_list, ignore_index= True)
```

## 6.1 INFERENCES FROM TESTING THE MODEL

### Considering the data is highly imbalanced, Accuracy is not the metrics to select the best performing model. Instead, we will look into the below metrics:
#### 1. Recall - What percent of all the fraud transactions, has our model been able to detect correctly. Higher this number, better is our model.
#### 2. False Positives - These are the authentic transactions, which the model has detected as fraud. Lower the number better is our model.
#### 3. Precision - What percent of the transactions predicted as Fraud by our model, are actually Fraud transactions. Again, we do not want too many transactions, incorrectly labelled as Fraud. Hence, higher the precision, better is our model. Considering the number of Fraud transactions are very low in our dataset, this percent will always be low and hence, we will consider False Positive count as a better metric.
#### 4. False Negatives - These are the fraud transactions, which the model has not been able to detect. Lower the number better is our model.


#### From the perspective of business, False Negatives are more costly. For the False Positives, the banks can always implement something like an additional verification. Hence,

#### Next for the finalized models, we will perform some hyper parameter tuning to see if we can further improve the performance of the models on our original test dataset.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RepeatedStratifiedKFold

# Hyper parameters for Logistic Regression
solvers = ['newton-cg', 'lbfgs', 'liblinear']
penalty = ['l1', 'l2']
c_values = [100, 10, 1.0, 0.1, 0.01]

# define grid search
grid_params = dict(solver=solvers,penalty=penalty,C=c_values)
cv = RepeatedStratifiedKFold(n_splits=7, n_repeats=3, random_state=24)

lr_grid_search = GridSearchCV(estimator=lr_allFtrs_direct, param_grid=grid_params, n_jobs=-1, cv = cv, scoring='recall')
lr_grid_result = lr_grid_search.fit(predictors_allFtrs_val, actual_val)
print("Logisitic Regression Best Recall: %f using %s" % (lr_grid_result.best_score_, lr_grid_result.best_params_))
```

```python
# Hyper parameters for Gaussian Naive Bayes
grid_params = {'var_smoothing': np.logspace(0,-9, num=100)}

nb_grid_search = GridSearchCV(estimator=nb_allFtrs_direct, param_grid=grid_params, n_jobs=-1, cv=cv, scoring='recall')
nb_grid_result = nb_grid_search.fit(predictors_allFtrs_val, actual_val)
print("Naive Bayes Best Recall: %f using %s" % (nb_grid_result.best_score_, nb_grid_result.best_params_))
```

```python
# Hyper parameters for Linear Discriminant Analysis
solvers = ['svd', 'lsqr', 'eigen']
grid_params = dict(solver=solvers)

lda_grid_search = GridSearchCV(estimator=lda_allFtrs_direct, param_grid=grid_params, n_jobs=-1, scoring='recall')
lda_grid_result = lda_grid_search.fit(predictors_allFtrs_val, actual_val)
print("Linear Discriminant Analysis Best Recall: %f using %s" % (lda_grid_result.best_score_, lda_grid_result.best_params_))
```

# 9.Test the Final Models

```python
final_scores = pd.DataFrame(columns = ['setName','ModelName','Features', 'Approach', 'Accuracy','Precision','Recall','confmatrix_TN','confmatrix_FP','confmatrix_FN','confmatrix_TP','fpr','tpr','auc'])
score_list = []

for (model, mdl_name, features, approach) in model_list_final:
    print(mdl_name)
    dict1 = {}
    accuracy, precision, recall, fpr, tpr, auc, cm  = fncValidate(model, mdl_name, predictors_allFtrs_val, actual_val)
    dict1.update({'setName': 'Validation',
                  'ModelName': mdl_name,
                  'Features': features,
                  'Approach': approach,
                  'Accuracy': accuracy,
                  'Precision': precision,
                  'Recall': recall,
                  'confmatrix_TN': cm[0,0],
                  'confmatrix_FP': cm[0,1],
                  'confmatrix_FN': cm[1,0],
                  'confmatrix_TP': cm[1,1],
                  'fpr': [fpr],
                  'tpr': [tpr],
                  'auc': auc})

    score_list.append(dict1)
final_scores = final_scores.append(score_list, ignore_index= True)
```
